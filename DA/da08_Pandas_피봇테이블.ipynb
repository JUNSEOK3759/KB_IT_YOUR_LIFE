### 1. Grouping

* groupby()  
> 데이타를 특정 기준으로 그룹핑 할때 사용(엑셀의 피봇 테이블과 유사)  
> 그룹핑을 한 후에는 반드시 통계함수를 적용한다.


'''
Grouping 이란 ??? --> 세분화, 잘게 쪼개는 것
'''

```python
# 성별로 그룹핑
# groupby 함수만으로는 아무것도 출력되지 않는다.
# 그룹핑 된 객체 자체만 리턴... DataFrameGroupBy
tips.groupby('sex')
```

```python
# 출력결과를 받아보려면.... DataFrameGroupBy객체에 통계함수를 적용해야 한다.

tips.groupby('sex').sum()
```

```python
#흡연, 비흡연 여부로 그룹핑.... 평균가를 출력

tips.groupby('smoker').mean()
```

#### 복수의 통계

* agg()

```python
# groupby를 이용해서 size 1.0, 2.0 .... 이 나왔는데, 이는 인덱스가 된다.

tips.groupby('size').agg(['sum', 'mean', 'count'])
```

#### 여러 개 컬럼으로 그룹핑

```python
# 흡연, 성별, 그룹핑....통계함수 mean
```

```python
tips.groupby(['smoker', 'sex']).mean()
```

```python
tips.groupby(['sex', 'day']).agg(['mean', 'sum'])
```

```python
tips.pivot_table(index = ['sex', 'day'], aggfunc =['mean', 'sum'])
```

### 2. 피봇테이블(엑셀의 그것과 동일)

**pivot_table() 피봇함수**는 DataFrame의 데이타를 Reshaping 하는 방법중 하나이다.

#### pivot_table 속성값

* values : 채우고자 하는 값
* index : 그룹핑 기준
* columns : 컬럼값

```python
tips.pivot_table(index = ['sex', 'day']).round(2)
```

```python
tips.pivot_table(values = ['total_bill', 'tip'], 
                index = ['sex', 'day'],
                columns = 'smoker')
```

```python
tips.pivot_table(['total_bill', 'tip'], 
                ['sex', 'day'],
                'smoker')
```

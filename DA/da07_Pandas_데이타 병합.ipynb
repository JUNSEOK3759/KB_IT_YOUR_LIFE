###  Pandas 데이타 병합하기


### Concat, Merge

    서로 다른 데이타프레임을 하나로 합치는 작업
    
    1) Concatenate 
    
      행 방향으로 연결됨 
      기본값이 axis=0 지정  
      
      이 경우에는 두 DataFrame이 서로 동일한 인덱스,컬럼을 가지고 있는 경우가 대부분  
      위+아래로 연결되는 방식이 기본이지만 좌,우로도 연결 가능하다.  
      
      outer join이 기본방식이다. 합집합
      key를 이용한 concat도 가능하다.
      
      
    2) Merge  
    
      두개의 DataFrame에 공통적으로 포함되어 있는 하나의 컬럼을 기준으로 합치는 방식       
      즉, 서로 다른 구성의 DataFrame이지만 공통된 key(컬럼)값을 가지고 있다면 병합 가능하다.  
      
      inner join이 기본방식. 교집합
      on 속성뒤 공통의 컬럼명
      how 속성뒤 조인기법을 적용함

### 1. Concat  
#### 행 방향으로 합치기

'''
concat은 열이 일치할 때 사용한다.
'''

```python
pd.concat([gas1, gas2])

#인덱스가 초기화되지 않았기에 총 행은 181714이지만 표 안에서는 91123밖에 안나온다.
```

```python
# concat 병합에서는 keys라는 옵션을 사용
pd.concat([gas1, gas2], keys = ['상반기', '하반기'])
```

#### 병합한 데이타를 gas에 입력

```python
gas = pd.concat([gas1, gas2])
```

```python
gas.iloc[90588:90593]

# 위에서 언급한 인덱스 초기화가 되지 않았기 때문이다.
```

#### index 리셋

```python
''' 
1. 인덱스는 잘 초기화 되어지지만, 인덱스 컬럼이 중복되어서 나온다. ...drop()
하면서 동시에 인덱스를 초기화 해줘야 한다.

2. 원본에는 적용이 되지 않는다.
'''

gas.reset_index()
gas.head(1)
```

```python
#1, 2번을 한 번에 해결
# drop = True하면 중복되는 인덱스가 사라지는 것을 볼 수 있다.
gas = gas.reset_index(drop = True)
```

```python
gas.head()
gas.tail()
```

#### 열 방향으로 합치기

```python
gas1 = gas.iloc[:, :4]
gas2 = gas.iloc[:, 4:]
```

```python
gas1.head()
gas2.head()
```

```python
pd.concat([gas1, gas2], axis = 1)
```

### 2. Merge

* 기본적으로 inner 방식으로 합쳐진다  (how='inner')  

상호간에 없는 데이타는 제외시키고   
서로 있는 데이타만 합쳐준다(inner방식 == 교집합).

```python
df1 = pd.DataFrame({
    '고객명':['박세리','이대호','손흥민','김연아', '마이클조던'],
    '생년월일':['1980-01-02','1982-02-22','1993-06-12','1988-10-16','1970-01-01'],
    '성별':['여자', '남자','남자','여자','남자']
})
df1
```

```python
df2= pd.DataFrame({
    '고객명':['김연아','박세리','손흥민','이대호','타이거우즈'],
    '연봉':['2000원','3000원','1500원','2500원','3500원']
})
df2
```

```python
'''
서로 상이한 데이터프레임에서 
고객명이라는 공통의 컬럼을 중심으로 
두 데이터프레임 집단을 병합..... merge

df1이라는 데이터프레임에서 df2 데이터프레임의 연봉정보를 가져오고 싶다.
'''

pd.merge(df1, df2)
# how의 디폴트값인 inner이 생략되었기에 결과가 교집합으로 나온다.
```

**how 옵션 4가지** 
* inner (default)
* reft
* right
* outer

```python
pd.merge(df1, df2, how = 'outer')
```

```python
pd.merge(df1, df2, how = 'left')

# 현업에서는 left를 많이 쓴다.
```

```python
pd.merge(df1, df2, how = 'right')
```

####  본질적으로 동일한 컬럼명 일치시키기

```python
df1 = pd.DataFrame({
    '이름':['박세리','이대호','손흥민','김연아', '마이클조던'],
    '생년월일':['1980-01-02','1982-02-22','1993-06-12','1988-10-16','1970-01-01'],
    '성별':['여자', '남자','남자','여자','남자']
})
df1

df2= pd.DataFrame({
    '고객명':['김연아','박세리','손흥민','이대호','타이거우즈'],
    '연봉':['2000원','3000원','1500원','2500원','3500원']
})
df2
```

```python
# 위의 '이름'과 '고객명'은 키의 이름이 다르긴 하지만 본질적으로 동일하다는 것을 알 수 있다.

pd.merge(df1, df2, left_on = '이름', right_on = '고객명')
```

```python
pd.merge(df1, df2, left_on = '이름', right_on = '고객명').drop('고객명', axis = 1)
#중복된 컬럼 제거
```

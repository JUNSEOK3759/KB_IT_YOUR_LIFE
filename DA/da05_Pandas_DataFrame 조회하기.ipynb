### 1. DataFrame - loc

판다스에서 가장 많이 사용하는 기능중 하나  

**인덱싱**, **슬라이싱**, **조건필터링** 기능까지 다 지원
이것만 잘 익혀두어도 많은 것을 손쉽게 할 수 있다.  

loc는 locate의 줄임말  
인덱싱이랑 비슷한데 다른 기능이 있기에 loc라 부른다.

```python
import numpy as np
import pandas as pd 
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
tips =pd.read_csv('../dataset/tips.csv')
tips.head(1)
```

```python
'''

numpy에서 했던 슬라이싱 방법이 안먹힌다.
tips[ 행기준 정렬, 열기준 정렬 ] -> error
'''
# tips[:, 0:2]


'''
dataframe에 loc를 적용하면 된다. df.loc
2.df.loc[ 행 기준 정렬 , 열 기준 정렬 ]
이 때, 행조건 앞, 뒤 전부다 포함
       열조건 앞, 뒤 전부다 포함
       
       처음에 헷갈리면 넣어보고 확인해가면서 사용하면 됩니다.
'''

tips.loc[:6, :'smoker'] # 원래는 [:6]이면 0 ~ 5까지 출력되는데, loc에서는 0 ~ 6까지 출력된다.
```

```python
tips.loc[6,'tip']
```

#### 조건 필터링

```python
tips
```

```python
'''
tip을 2불보다 많이 지불한 사람의 데이터...

> 한방에 코드를 작성
> step by step

1. 조건을 먼저 정의
2. 정의한 조건을 변수에 할당 ...condition 이라고 변수 지정
3. df.loc[condition]
'''


condition = (tips['tip'] < 2)
tips.loc[condition]
```

#### 문제 1

고객으로부터 새로운 분석 요구조건이 들어왔습니다  
팁을 2불 이상 낸 사람중에서 남자들만 검색해달라.  

어떻게 작성해야 할까?

```python
condition1 = (tips['tip'] >= 2)
condition2 = (tips['sex'] == 'Male')
condition3 = (tips['size'] == 2)
condition = tips.loc[condition1 & condition2 & condition3]
```

#### 문제 2. 
기존의 조건을 그대로 유지하는데  
컬럼이 너무 많아서 보기가 힘이 드니  
total_bill, tip만 뽑아서 보고 싶다  

어떻게 작성해야 할까?

```python
tips.loc[condition1 & condition2& condition3, 'total_bill' : 'tip']
tips.loc[condition1 & condition2 & condition3, : 'tip']
tips.loc[condition1 & condition2 & condition3, ['tip', 'smoker', 'size']] # 열거형으로 내가 원하는 컬럼 가져오기 가능
```

```python
tips.loc[condition1 & condition2 & condition3].head(3)
#이건 loc로 추출한 것이다.
```

```python
tips[condition1 & condition2 & condition3].head(3)
#이건 boolean indexing으로 추출한 것이다.... 권장하지 않음
```

```python
tips[condition1 & condition2 & condition3]['time'].head(3)

tips.loc[condition1 & condition2 & condition3, 'time'].head(3)
```

```python
tips[condition1]['tip'].head(3)

tips.loc[condition1, 'tip'].head(3)
```

```python
'''
boolean indexing으로 값 대입하면 잘 안나온다.
이게 사실 위험한 작업이다. 조건부 필터링을 걸어서 특정한 컬럼을 새로운 다른 값으로
채우는 일을 많이 한다. 이 때, 우리는 새로운 값이 잘 대입되었다고 생각하지만 
실상은 그렇지 않은 결과가 많다. 분석에 치명적인 실수를 야기하는 부분이다.

결론 : loc를 사용하라.

'''

tips[condition1]['tip'] = 2

# 그저 경고일 뿐이다.
```

```python
tips.loc[condition1, 'tip'].head(3)
```

```python
# .loc를 사용해서 해당조건일 때, 2로 할당
tips.loc[condition1, 'tip'] = 2 # 2 달러 이상일 때, 전부 다 2달러로 재할당
tips.head(10)
```

### 연습문제 1.

아래 조건을 만족하는 코드를 입력하세요  

* 남성, 흡연자 손님을 필터링
* tip을 많이 낸 순서대로 정렬
* 상위 10개를 출력

```python
cond1 = (tips['sex'] == 'Male')
cond2 = (tips['smoker'] == 'Yes')
tips.loc[cond1 & cond2].sort_values(by = 'tip', ascending = False).head(10)
```

### 연습문제 2

아래 조건을 만족하는 코드를 작성하세요.

* 팁을 5 이상, 10 이하로 낸 손님을 필터링
* size가 3인 이하 손님을 필터링
* 컬럼은 totla_bill, sex, time만 나오도록 출력



```python
cond1 = (tips['tip'] >= 5) & (tips['tip'] <= 10)
cond2 = (tips['size'] <= 3)
tips.loc[cond1&cond2, ['total_bill', 'sex', 'time']]
```

### 2. DataFrame - iloc

loc앞에 i만 붙었다.  
i는 index의 i  
loc와 다르게 iloc는 label이 들어갈 수 없다.  
무조건 index로만 검색하는 기능  

조건 필터링할때는 안쓰이고 주로 슬라이싱 할때 많이 사용됨  Numpy의 슬라이싱과 완전 동일한 기법

```python
tips.head(1)
```

```python
tips.columns
```

```python
# indexing 원리이기 때문에 loc처럼 :101이라고 해도 101번까지 나오는게 아니라 101개가 출력되어 
#0부터 100까지 총 101개가 출력된다.
# 즉, 마지막 인덱싱은 포함하지 않는다.

tips.iloc[: 101, [1,3,5]]


'''

실무데이터는 컬럼의 갯수가 1000개가 넘는 것들이 많다.
이럴 때는 iloc가 편하다.

실무에서 보통은 iloc[0 : 100]으로 가져와서 전처리하고, 
이후에 또 iloc[101 : 200]으로 가져와서 전처리하고 ... 이와 같은 방식이다.

큰 데이터를 100개 단위로 전처리하는 방식으로 하고 이후에 merge하는데, 이 때 iloc를 사용한다.

'''
```
